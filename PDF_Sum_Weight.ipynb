```python
df = tables[0]
df
```

```python
columns = df.columns.values
columns
```

```python
df.loc[len(df)] = pd.Series(columns,df.columns, name='new_row')
df
```

```python
size = ['L45X4','L50X4','L50X6','L60X4','L60X5','L65X6','L70X6','L75X6','L80X6','L90X6','L90X77','L90X10','L100X7','L100X10','L120X8','L130X9','L130X12','L150X10','L150X12','L150X157','L150X19','L175X12','L175X15','L200X15','L200X20','L200X25','L250X25','L250X35']
```

```python
def rename_column(dataframe):
    org_namelist = list(dataframe.columns)

    # rename each list
    for i in range(len(org_namelist)):
      if org_namelist[i] in size:
        org_namelist[i] = 'size'
        org_namelist[i+1] = 'length'
        org_namelist[i+2] = 'number'
        org_namelist[i+3] = 'the mass of each'
        org_namelist[i+4] = 'number * mass of each'
        break
      else:
        org_namelist[i] = i

    # change the dataframe lable
    dataframe = dataframe.set_axis(org_namelist, axis = 1, inplace = False)

    return dataframe
```

```python
df = rename_column(df)
df.tail(10)
```

```python
df
```

```python
df.dropna(subset=['size'],inplace=True)
```

```python
df
```

```python
abnormal_row_list = df.iloc[39]
df.drop(39,inplace=True)
for i in [0,1]:
  df.loc[len(df)] = pd.Series(abnormal_row_list.str.split(' ').str.get(i))
```

```python
df
```

```python
df.dropna(subset=['size'],inplace=True)
```

```python
column_list = ['length', 'number', 'the mass of each', 'number * mass of each']
for i in range(4):
  df[column_list[i]] = df[column_list[i]].astype(float)
```

```python
sum_weigh_df = df.groupby(['size']).sum()
sum_weigh_df.rename(columns={'number * mass of each':'total_mass'},inplace=True)
```
